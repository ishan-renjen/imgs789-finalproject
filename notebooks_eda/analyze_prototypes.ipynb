```python
superclass_map = {
    0:'vehicle', 
    1:'vehicle',
    2:'outdoor', 
    3:'outdoor',
    4:'animal', 
    5:'animal', 
    6:'animal',
    7:'accessory',
    8:'sports', 
    9:'sports',
    10:'kitchen', 
    11:'kitchen',
    12:'food', 
    13:'food',
    14:'furniture', 
    15:'furniture',
    16:'electronic',
    17:'appliance', 
    18:'appliance',
    19:'indoor',
    #end task 1
    20:'person', 
    21:'vehicle',
    22:'vehicle', 
    23:'outdoor',
    24:'animal', 
    25:'animal', 
    26:'accessory',
    27:'sports',
    28:'sports', 
    29:'kitchen',
    30:'food', 
    31:'food',
    32:'food', 
    33:'furniture',
    34:'furniture', 
    35:'electronic',
    36:'electronic',
    37:'appliance', 
    38:'indoor',
    39:'indoor',
    #end task 2
    40: 'vehicle',
    41: 'vehicle',
    42: 'outdoor',
    43: 'animal',
    44: 'animal',
    45: 'animal',
    46: 'accessory',
    47: 'sports',
    48: 'sports',
    49: 'sports',
    50: 'kitchen',
    51: 'kitchen',
    52: 'food',
    53: 'food',
    54: 'kitchen',
    55: 'electronic',
    56: 'electronic',
    57: 'appliance',
    58: 'indoor',
    59: 'indoor',
    #end task 3
    60: 'vehicle',
    61: 'vehicle',
    62: 'outdoor',
    63: 'animal',
    64: 'animal',
    65: 'accessory',
    66: 'accessory',
    67: 'sports',
    68: 'sports',
    69: 'sports',
    70: 'kitchen',
    71: 'kitchen',
    72: 'food',
    73: 'food',
    74: 'food',
    75: 'kitchen',
    76: 'electronic',
    77: 'appliance',
    78: 'indoor',
    79: 'indoor',
}

def build_class_prototypes(embeddings, labels, k=10):
    prototypes = dict()

    for class_label in np.unique(labels):
        class_embeddings = embeddings[labels == class_label]

        if len(class_embeddings) <= k:
            prototypes[int(class_label)] = [class_embeddings.mean(axis=0)]
            continue

        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(class_embeddings)
        prototypes[int(class_label)] = [center for center in kmeans.cluster_centers_]

    return prototypes

def build_superclass_prototypes(class_prototypes: dict, superclass_map=superclass_map):
    superclass_prototypes = dict()

    for class_id, prototypes in class_prototypes.items():
        superclass = superclass_map[class_id]
        superclass_prototypes.setdefault(superclass, []).extend(prototypes)

    return superclass_prototypes

def build_class_embedding_dict(embeddings, labels):
    class_embeddings = {}
    for class_id in np.unique(labels):
        class_embeddings[int(class_id)] = np.asarray(embeddings[labels == class_id], dtype=np.float64)
    return class_embeddings

def compute_class_spread_stats(class_embedding_dict, reg=1e-6):
    stats = {}

    for class_id, X in class_embedding_dict.items():
        X = np.asarray(X, dtype=np.float64)
        mu = X.mean(axis=0)
        centered = X - mu

        dists = np.linalg.norm(centered, axis=1)

        if len(X) > 1:
            var = X.var(axis=0, ddof=1) + reg
            cov = np.cov(X, rowvar=False)
        else:
            var = np.full(X.shape[1], reg, dtype=np.float64)
            cov = np.eye(X.shape[1], dtype=np.float64) * reg

        stats[class_id] = {
            "n": len(X),
            "mean_radius": float(dists.mean()),
            "p95_radius": float(np.percentile(dists, 95)),
            "trace_cov": float(np.trace(cov)),
            "mean_diag_var": float(var.mean()),
        }

    return stats
```

```python
def get_regularized_inverse_covariance(set1_array, set2_array, reg=1e-3):
    set1_array = np.asarray(set1_array, dtype=np.float64)
    set2_array = np.asarray(set2_array, dtype=np.float64)

    all_points = np.vstack([set1_array, set2_array])

    # If there is only one point total, covariance is not defined
    if all_points.shape[0] < 2:
        d = all_points.shape[1]
        return np.eye(d, dtype=np.float64)

    cov = np.cov(all_points, rowvar=False)
    cov = np.atleast_2d(cov)

    d = cov.shape[0]
    cov = cov + reg * np.eye(d, dtype=np.float64)

    # pinv is safer than inv in high dimensions
    VI = np.linalg.pinv(cov)
    return VI

def get_matched_prototypes(class_prototype_set1, class_prototype_set2, metric='euclidean', reg=1e-3, class_embedding_set1=None, class_embedding_set2=None):
    shared_classes = sorted(set(class_prototype_set1.keys()) & set(class_prototype_set2.keys()))

    if len(shared_classes) == 0:
        return {}, np.nan, {}

    matched_pairs_by_class = {}
    class_drifts = {}

    for class_id in shared_classes:
        set1_array = np.asarray(class_prototype_set1[class_id], dtype=np.float64)
        set2_array = np.asarray(class_prototype_set2[class_id], dtype=np.float64)

        if set1_array.ndim == 1:
            set1_array = set1_array.reshape(1, -1)
        if set2_array.ndim == 1:
            set2_array = set2_array.reshape(1, -1)

        if metric == 'euclidean':
            cost_matrix = cdist(set1_array, set2_array, metric='euclidean')

        elif metric == 'mahalanobis':
            if class_embedding_set1 is None or class_embedding_set2 is None:
                raise ValueError("Mahalanobis requires raw class embedding sets.")

            # VI = get_classwise_inverse_covariance_from_raw_embeddings(
            VI = get_classwise_inverse_diag_cov_from_raw_embeddings(
                class_id,
                class_embedding_set1,
                class_embedding_set2,
                reg=reg
            )
            cost_matrix = cdist(set1_array, set2_array, metric='mahalanobis', VI=VI)

        else:
            raise ValueError(f"Unsupported metric: {metric}")

        row_ind, col_ind = scipy.linear_sum_assignment(cost_matrix)
        matched_pairs_by_class[class_id] = list(zip(row_ind.tolist(), col_ind.tolist()))
        class_drifts[class_id] = float(cost_matrix[row_ind, col_ind].mean())

    overall_drift = float(np.mean(list(class_drifts.values())))
    return matched_pairs_by_class, overall_drift, class_drifts

def get_classwise_inverse_covariance_from_raw_embeddings(class_id, class_embedding_set1, class_embedding_set2, reg=1e-3):
    emb1 = np.asarray(class_embedding_set1[class_id], dtype=np.float64)
    emb2 = np.asarray(class_embedding_set2[class_id], dtype=np.float64)

    all_embeddings = np.vstack([emb1, emb2])

    if all_embeddings.shape[0] < 2:
        d = all_embeddings.shape[1]
        return np.eye(d, dtype=np.float64)

    lw = LedoitWolf()
    lw.fit(all_embeddings)
    cov = lw.covariance_ + reg * np.eye(all_embeddings.shape[1], dtype=np.float64)

    VI = np.linalg.pinv(cov)
    return VI

def get_classwise_inverse_diag_cov_from_raw_embeddings(class_id, class_embedding_set1, class_embedding_set2, reg=1e-3):
    emb1 = np.asarray(class_embedding_set1[class_id], dtype=np.float64)
    emb2 = np.asarray(class_embedding_set2[class_id], dtype=np.float64)

    all_embeddings = np.vstack([emb1, emb2])

    if all_embeddings.shape[0] < 2:
        d = all_embeddings.shape[1]
        return np.eye(d, dtype=np.float64)

    var = np.var(all_embeddings, axis=0, ddof=1) + reg
    VI = np.diag(1.0 / var)
    return VI

def build_drift_matrix(task_prototype_sets, metric='euclidean', reg=1e-3, task_embedding_sets=None):
    n = len(task_prototype_sets)
    drift_matrix = np.zeros((n, n), dtype=np.float64)

    all_matches = [[None for _ in range(n)] for _ in range(n)]
    all_class_drifts = [[None for _ in range(n)] for _ in range(n)]

    for i in range(n):
        for j in range(n):
            matched_pairs_by_class, overall_drift, class_drifts = get_matched_prototypes(
                task_prototype_sets[i],
                task_prototype_sets[j],
                metric=metric,
                reg=reg,
                class_embedding_set1=None if task_embedding_sets is None else task_embedding_sets[i],
                class_embedding_set2=None if task_embedding_sets is None else task_embedding_sets[j],
            )
            drift_matrix[i, j] = overall_drift
            all_matches[i][j] = matched_pairs_by_class
            all_class_drifts[i][j] = class_drifts

    return drift_matrix, all_matches, all_class_drifts
```

```python
def get_single_class_drift(class_id,
                           task_prototype_set1,
                           task_prototype_set2,
                           metric='euclidean',
                           reg=1e-3,
                           task_embedding_set1=None,
                           task_embedding_set2=None):
    if class_id not in task_prototype_set1 or class_id not in task_prototype_set2:
        return np.nan

    set1_array = np.asarray(task_prototype_set1[class_id], dtype=np.float64)
    set2_array = np.asarray(task_prototype_set2[class_id], dtype=np.float64)

    if set1_array.ndim == 1:
        set1_array = set1_array.reshape(1, -1)
    if set2_array.ndim == 1:
        set2_array = set2_array.reshape(1, -1)

    if metric == 'euclidean':
        cost_matrix = cdist(set1_array, set2_array, metric='euclidean')

    elif metric == 'mahalanobis':
        if task_embedding_set1 is None or task_embedding_set2 is None:
            raise ValueError("Mahalanobis requires raw class embedding sets.")

        VI = get_classwise_inverse_diag_cov_from_raw_embeddings(
            class_id,
            task_embedding_set1,
            task_embedding_set2,
            reg=reg
        )
        cost_matrix = cdist(set1_array, set2_array, metric='mahalanobis', VI=VI)

    else:
        raise ValueError(f"Unsupported metric: {metric}")

    row_ind, col_ind = scipy.linear_sum_assignment(cost_matrix)
    return float(cost_matrix[row_ind, col_ind].mean())

def build_per_class_task_drift_matrix(class_id,
                                      task_prototype_sets,
                                      metric='euclidean',
                                      reg=1e-3,
                                      task_embedding_sets=None):
    n = len(task_prototype_sets)
    drift_matrix = np.full((n, n), np.nan, dtype=np.float64)

    for i in range(n):
        for j in range(n):
            drift_matrix[i, j] = get_single_class_drift(
                class_id,
                task_prototype_sets[i],
                task_prototype_sets[j],
                metric=metric,
                reg=reg,
                task_embedding_set1=None if task_embedding_sets is None else task_embedding_sets[i],
                task_embedding_set2=None if task_embedding_sets is None else task_embedding_sets[j],
            )

    return drift_matrix
```
