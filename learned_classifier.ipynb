```python
class PrototypicalNetwork(torch.nn.Module):
    def __init__(self, input_size, output_size, dropout=0.0):
        super().__init__()
        hidden_size_1 = int(output_size + (input_size - output_size / 2))
        hidden_size_2 = int(output_size + (input_size - output_size / 4))
        hidden_size_3 = int(output_size + (input_size - output_size / 8))

        self.layer1 = torch.nn.Linear(input_size, hidden_size_1)
        self.layer2 = torch.nn.Linear(hidden_size_1, hidden_size_2)
        self.layer3 = torch.nn.Linear(hidden_size_2, hidden_size_3)

        self.tanh = torch.nn.Tanh()
        self.dropout = torch.nn.Dropout(dropout)

        self.layer_out = torch.nn.Linear(hidden_size_3, output_size)

    def forward(self, x):
        # Define how data passes through the layers
        x = self.layer1(x)
        x = self.tanh(x)
        x = self.dropout(x)
        x = self.layer2(x)
        x = self.tanh(x)
        x = self.dropout(x)
        x = self.layer3(x)
        x = self.tanh(x)
        x = self.dropout(x)
        x = self.layer_out(x)
        return x
```

```python
def l2_normalize(x, eps=1e-12):
    pca = PCA(n_components=24, random_state=42)
    class_embeddings_reduced = pca.fit_transform(x)
    return x / (np.linalg.norm(class_embeddings_reduced, axis=1, keepdims=True) + eps)

def make_superclass_ids(labels, superclass_map):
    superclass_names = sorted({superclass_map[int(y)] for y in np.unique(labels)})
    sc_to_id = {name: i for i, name in enumerate(superclass_names)}
    y_sc = np.array([sc_to_id[superclass_map[int(y)]] for y in labels], dtype=np.int64)
    return y_sc, sc_to_id

def make_split(embeddings, labels, holdout_class):
    train_mask = labels != holdout_class
    test_mask = labels == holdout_class

    X_train = embeddings[train_mask]
    y_train = labels[train_mask]

    X_test = embeddings[test_mask]
    y_test = labels[test_mask]

    return X_train, y_train, X_test, y_test

def train_superclass_classifier(
    X_train,
    y_train_sc,
    X_val,
    y_val_sc,
    num_superclasses,
    model_type="mlp",
    lr=1e-3,
    batch_size=256,
    epochs=50,
    device="cuda" if torch.cuda.is_available() else "cpu"
):
    X_train_t = torch.tensor(X_train, dtype=torch.float32)
    y_train_t = torch.tensor(y_train_sc, dtype=torch.long)
    X_val_t = torch.tensor(X_val, dtype=torch.float32)
    y_val_t = torch.tensor(y_val_sc, dtype=torch.long)

    train_loader = DataLoader(
        TensorDataset(X_train_t, y_train_t),
        batch_size=batch_size,
        shuffle=True
    )
    val_loader = DataLoader(
        TensorDataset(X_val_t, y_val_t),
        batch_size=batch_size,
        shuffle=False
    )

    if model_type == "linear":
        model = PrototypicalNetwork(input_size=X_train.shape[1], dropout = 0.1, output_size=num_superclasses)
    else:
        model = PrototypicalNetwork(input_size=X_train.shape[1], dropout = 0.1, output_size=num_superclasses)

    model = model.to(device)

    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)

    best_val_acc = -1.0
    best_state = None

    for epoch in range(epochs):
        model.train()
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)

            optimizer.zero_grad()
            logits = model(xb)
            loss = criterion(logits, yb)
            loss.backward()
            optimizer.step()

        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                preds = model(xb).argmax(dim=1)
                correct += (preds == yb).sum().item()
                total += yb.size(0)

        val_acc = correct / max(total, 1)

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}

    model.load_state_dict(best_state)
    model = model.to(device)
    return model, best_val_acc

def evaluate_model(model, X_test, y_test_sc, device="cuda" if torch.cuda.is_available() else "cpu"):
    model.eval()
    X_test_t = torch.tensor(X_test, dtype=torch.float32).to(device)
    y_test_t = torch.tensor(y_test_sc, dtype=torch.long).to(device)

    with torch.no_grad():
        logits = model(X_test_t)
        preds = logits.argmax(dim=1)

    acc = (preds == y_test_t).float().mean().item()
    return acc, preds.cpu().numpy()
```

```python
def run_superclass_mlp(embeddings, labels, holdout_class, superclass_map=superclass_map, model_type="mlp"):
    X_train, y_train_fg, X_test, y_test_fg = make_split(embeddings, labels, holdout_class)

    y_train_sc, sc_to_id = make_superclass_ids(y_train_fg, superclass_map)
    y_test_sc = np.array([sc_to_id[superclass_map[int(y)]] for y in y_test_fg], dtype=np.int64)

    X_tr, X_val, y_tr, y_val = train_test_split(
        X_train, y_train_sc,
        test_size=0.2,
        random_state=42,
        stratify=y_train_sc)

    model, best_val_acc = train_superclass_classifier(
        X_tr, y_tr,
        X_val, y_val,
        num_superclasses=len(sc_to_id),
        model_type=model_type)

    test_acc, preds = evaluate_model(model, X_test, y_test_sc)
    return test_acc, best_val_acc, preds, sc_to_id
```

```python
x = l2_normalize(t1_euclidean_embeddings_known)
y = t1_euclidean_labels_known

for idx in range(0, 20):
    train_labels = y[y != idx]
    train_superclasses = {superclass_map[int(c)] for c in np.unique(train_labels)}
    test_superclass = superclass_map[idx]

    if test_superclass not in train_superclasses:
        continue

    test_acc, best_val_acc, preds, sc_to_id = run_superclass_mlp(x, y, idx)
    print(f"held-out class {idx}, test accuracy: {test_acc*100}%")
```
