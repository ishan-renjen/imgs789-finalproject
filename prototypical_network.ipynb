```python
def l2_normalize(x, eps=1e-12):
    return x / (np.linalg.norm(x, axis=1, keepdims=True) + eps)

def build_class_prototypes(embeddings, labels, k=5):
    prototypes = dict()

    for class_label in np.unique(labels):
        class_embeddings = embeddings[labels == class_label]

        if len(class_embeddings) <= k:
            prototypes[int(class_label)] = [class_embeddings.mean(axis=0)]
            continue
        
        pca = PCA(n_components=32, random_state=42)
        class_embeddings_reduced = pca.fit_transform(class_embeddings)
        clusterer = hdbscan.HDBSCAN(min_cluster_size=k, min_samples=1, metric='euclidean', cluster_selection_method='eom', prediction_data=False)

        class_protos = []
        cluster_ids = clusterer.fit_predict(class_embeddings_reduced)
        unique_clusters = [cid for cid in np.unique(cluster_ids) if cid != -1]

        if len(unique_clusters) == 0:
            print(f"length of clusters for class label {class_label} = 0")
            proto = class_embeddings.mean(axis=0)
            proto = proto / (np.linalg.norm(proto) + 1e-12)
            prototypes[int(class_label)] = [proto]
            continue

        for cid in unique_clusters:
            cluster_points = class_embeddings[cluster_ids == cid]
            proto = cluster_points.mean(axis=0)
            proto = proto / (np.linalg.norm(proto) + 1e-12)
            class_protos.append(proto)

        prototypes[int(class_label)] = class_protos

    return prototypes

def build_class_prototypes_kmeans(embeddings, labels, k=10):
    prototypes = dict()

    for class_label in np.unique(labels):
        class_embeddings = embeddings[labels == class_label]

        if len(class_embeddings) <= k:
            prototypes[int(class_label)] = [class_embeddings.mean(axis=0)]
            continue

        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(class_embeddings)
        prototypes[int(class_label)] = [center for center in kmeans.cluster_centers_]

    return prototypes

def build_superclass_prototypes(class_prototypes: dict, superclass_map=superclass_map):
    superclass_prototypes = dict()

    for class_id, prototypes in class_prototypes.items():
        superclass = superclass_map[class_id]
        superclass_prototypes.setdefault(superclass, []).extend(prototypes)

    return superclass_prototypes

def cosine_loss(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    
    similarity = dot_product / (norm_a * norm_b)
    return 1 - similarity

def predict_with_superclass(query, prototypes):
    pred = None
    lowest_loss = float('inf')

    for superclass, prototypes in prototypes.items():
        for prototype in prototypes:
            loss = cosine_loss(query, prototype)
            if loss < lowest_loss:
                lowest_loss = loss
                pred = superclass
    
    return pred, lowest_loss

def predict_with_superclass_avg_cos(query, prototypes):
    best_superclass = None
    best_score = float("inf")

    for superclass, proto_list in prototypes.items():
        losses = [cosine_loss(query, proto) for proto in proto_list]
        avg_loss = np.mean(losses)

        if avg_loss < best_score:
            best_score = avg_loss
            best_superclass = superclass

    return best_superclass, best_score

def predict_with_superclass_avg_l1(query, prototypes):
    best_superclass = None
    best_score = float("inf")

    for superclass, proto_list in prototypes.items():
        losses = [np.linalg.norm(query - proto, ord=1) for proto in proto_list]
        avg_loss = np.mean(losses)

        if avg_loss < best_score:
            best_score = avg_loss
            best_superclass = superclass

    return best_superclass, best_score

def predict_with_superclass_vote(query, prototypes, k=5):
    neighbors = []

    for superclass, proto_list in prototypes.items():
        for proto in proto_list:
            loss = cosine_loss(query, proto)
            neighbors.append((loss, superclass))

    neighbors.sort(key=lambda x: x[0])
    topk = neighbors[:k]

    votes = {}
    for _, superclass in topk:
        votes[superclass] = votes.get(superclass, 0) + 1

    pred = max(votes.items(), key=lambda x: x[1])[0]
    return pred, topk
```

```python
class_holdouts = {0, 1, 2, 3, 4, 5, 6, 8, 9, 10, 11, 12, 13, 14, 15, 17, 18}
for class_label in class_holdouts:
    x = l2_normalize(t1_euclidean_embeddings_known)
    y = t1_euclidean_labels_known

    holdout_mask = np.isin(t1_euclidean_labels_known, [class_label])
    t1_euclidean_labels_holdout = t1_euclidean_labels_known[holdout_mask]
    t1_euclidean_embeddings_holdout = x[holdout_mask]

    t1_euclidean_training_mask = t1_euclidean_labels_known != class_label
    t1_euclidean_training_labels = t1_euclidean_labels_known[t1_euclidean_training_mask]
    t1_euclidean_training_emb = x[t1_euclidean_training_mask]

    class_protos = build_class_prototypes(t1_euclidean_training_emb, t1_euclidean_training_labels)
    superclass_protos = build_superclass_prototypes(class_protos, superclass_map)

    correct = 0
    total = len(t1_euclidean_embeddings_holdout)

    for idx in range(total):
        pred, sim = predict_with_superclass(
            t1_euclidean_embeddings_holdout[idx],
            superclass_protos
        )
        true_sc = superclass_map[int(t1_euclidean_labels_holdout[idx])]
        if pred == true_sc:
            correct += 1

    print(f"accuracy on held-out class {class_label}:", correct / total)
```

```python
class PrrototypicalNetwork(torch.nn.Module):
    def __init__(self, input_size, hidden_size, output_size, dropout=0.0):
        super(PrrototypicalNetwork).__init__()
        self.layer1 = torch.nn.Linear(input_size, hidden_size)
        self.relu = torch.nn.Tanh()
        self.dropout = torch.nn.Dropout(dropout)
        self.layer2 = torch.nn.Linear(hidden_size, output_size)

    def forward(self, x):
        # Define how data passes through the layers
        x = self.layer1(x)
        x = self.relu(x)
        self.dropout(x)
        x = self.layer2(x)
        return x
```
