```python
def l2_normalize(x, eps=1e-12):
    return x / (np.linalg.norm(x, axis=1, keepdims=True) + eps)

def build_class_prototypes(embeddings, labels, k=5):
    prototypes = dict()

    for class_label in np.unique(labels):
        class_embeddings = embeddings[labels == class_label]

        if len(class_embeddings) <= k:
            prototypes[int(class_label)] = [class_embeddings.mean(axis=0)]
            continue
        
        pca = PCA(n_components=16, random_state=42)
        class_embeddings_reduced = pca.fit_transform(class_embeddings)
        clusterer = hdbscan.HDBSCAN(min_cluster_size=k, min_samples=1, metric='euclidean', cluster_selection_method='eom', prediction_data=False)

        class_protos = []
        cluster_ids = clusterer.fit_predict(class_embeddings_reduced)
        unique_clusters = [cid for cid in np.unique(cluster_ids) if cid != -1]

        if len(unique_clusters) == 0:
            print(f"length of clusters for class label {class_label} = 0")
            proto = class_embeddings.mean(axis=0)
            proto = proto / (np.linalg.norm(proto) + 1e-12)
            prototypes[int(class_label)] = [proto]
            continue

        for cid in unique_clusters:
            cluster_points = class_embeddings[cluster_ids == cid]
            proto = cluster_points.mean(axis=0)
            proto = proto / (np.linalg.norm(proto) + 1e-12)
            class_protos.append(proto)

        prototypes[int(class_label)] = class_protos

    return prototypes

def build_class_prototypes_kmeans(embeddings, labels, k=10):
    prototypes = dict()

    for class_label in np.unique(labels):
        class_embeddings = embeddings[labels == class_label]

        if len(class_embeddings) <= k:
            prototypes[int(class_label)] = [class_embeddings.mean(axis=0)]
            continue

        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(class_embeddings)
        prototypes[int(class_label)] = [center for center in kmeans.cluster_centers_]

    return prototypes

def build_superclass_prototypes(class_prototypes: dict, superclass_map=superclass_map):
    superclass_prototypes = dict()

    for class_id, prototypes in class_prototypes.items():
        superclass = superclass_map[class_id]
        superclass_prototypes.setdefault(superclass, []).extend(prototypes)

    return superclass_prototypes

def cosine_loss(a, b):
    dot_product = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    
    similarity = dot_product / (norm_a * norm_b)
    return 1 - similarity

def predict_with_superclass(query, prototypes):
    pred = None
    lowest_loss = float('inf')

    for superclass, prototypes in prototypes.items():
        for prototype in prototypes:
            loss = cosine_loss(query, prototype)
            if loss < lowest_loss:
                lowest_loss = loss
                pred = superclass
    
    return pred, lowest_loss

def predict_with_superclass_avg_cos(query, prototypes):
    best_superclass = None
    best_score = float("inf")

    for superclass, proto_list in prototypes.items():
        losses = [cosine_loss(query, proto) for proto in proto_list]
        avg_loss = np.mean(losses)

        if avg_loss < best_score:
            best_score = avg_loss
            best_superclass = superclass

    return best_superclass, best_score

def predict_with_superclass_avg_l1(query, prototypes):
    best_superclass = None
    best_score = float("inf")

    for superclass, proto_list in prototypes.items():
        losses = [np.linalg.norm(query - proto, ord=1) for proto in proto_list]
        avg_loss = np.mean(losses)

        if avg_loss < best_score:
            best_score = avg_loss
            best_superclass = superclass

    return best_superclass, best_score

def predict_with_superclass_vote(query, prototypes, k=5):
    neighbors = []

    for superclass, proto_list in prototypes.items():
        for proto in proto_list:
            loss = cosine_loss(query, proto)
            neighbors.append((loss, superclass))

    neighbors.sort(key=lambda x: x[0])
    topk = neighbors[:k]

    votes = {}
    for _, superclass in topk:
        votes[superclass] = votes.get(superclass, 0) + 1

    pred = max(votes.items(), key=lambda x: x[1])[0]
    return pred, topk

def get_regularized_inverse_covariance(points, reg=1e-3):
    points = np.asarray(points, dtype=np.float64)
    if len(points) < 2:
        d = points.shape[1]
        return np.eye(d, dtype=np.float64)

    lw = LedoitWolf().fit(points)
    cov = lw.covariance_
    return np.linalg.pinv(cov)

def mahalanobis_distance(x, mean, VI):
    diff = x - mean
    return float(diff @ VI @ diff)

def build_superclass_stats(embeddings, labels, superclass_map, reg=1e-3):
    stats = {}

    superclasses = sorted({superclass_map[int(y)] for y in np.unique(labels)})
    for sc in superclasses:
        mask = np.array([superclass_map[int(y)] == sc for y in labels])
        X_sc = embeddings[mask]

        mu = X_sc.mean(axis=0)
        VI = get_regularized_inverse_covariance(X_sc, reg=reg)

        stats[sc] = {"mean": mu, "VI": VI}

    return stats

def predict_with_superclass_mahalanobis(query, stats):
    best_sc = None
    best_dist = float("inf")

    for sc, s in stats.items():
        dist = mahalanobis_distance(query, s["mean"], s["VI"])
        if dist < best_dist:
            best_dist = dist
            best_sc = sc

    return best_sc, best_dist
```

```python
class PrototypicalNetwork(torch.nn.Module):
    def __init__(self, input_size, output_size, dropout=0.0):
        super().__init__()
        hidden_size_1 = int(output_size + (input_size - output_size / 2))
        hidden_size_2 = int(output_size + (input_size - output_size / 4))
        hidden_size_3 = int(output_size + (input_size - output_size / 8))

        self.layer1 = torch.nn.Linear(input_size, hidden_size_1)
        self.layer2 = torch.nn.Linear(hidden_size_1, hidden_size_2)
        self.layer3 = torch.nn.Linear(hidden_size_2, hidden_size_3)

        self.tanh = torch.nn.Tanh()
        self.dropout = torch.nn.Dropout(dropout)

        self.layer_out = torch.nn.Linear(hidden_size_3, output_size)

    def forward(self, x):
        # Define how data passes through the layers
        x = self.layer1(x)
        x = self.tanh(x)
        x = self.dropout(x)
        x = self.layer2(x)
        x = self.tanh(x)
        x = self.dropout(x)
        x = self.layer3(x)
        x = self.tanh(x)
        x = self.dropout(x)
        x = self.layer_out(x)
        return x
```

```python
def make_superclass_ids(labels, superclass_map):
    superclass_names = sorted({superclass_map[int(y)] for y in np.unique(labels)})
    sc_to_id = {name: i for i, name in enumerate(superclass_names)}
    y_sc = np.array([sc_to_id[superclass_map[int(y)]] for y in labels], dtype=np.int64)
    return y_sc, sc_to_id

def make_split(embeddings, labels, holdout_class):
    train_mask = labels != holdout_class
    test_mask = labels == holdout_class

    X_train = embeddings[train_mask]
    y_train = labels[train_mask]

    X_test = embeddings[test_mask]
    y_test = labels[test_mask]

    return X_train, y_train, X_test, y_test

def train_superclass_classifier(
    X_train,
    y_train_sc,
    X_val,
    y_val_sc,
    num_superclasses,
    model_type="mlp",
    lr=1e-3,
    batch_size=256,
    epochs=50,
    device="cuda" if torch.cuda.is_available() else "cpu"
):
    X_train_t = torch.tensor(X_train, dtype=torch.float32)
    y_train_t = torch.tensor(y_train_sc, dtype=torch.long)
    X_val_t = torch.tensor(X_val, dtype=torch.float32)
    y_val_t = torch.tensor(y_val_sc, dtype=torch.long)

    train_loader = DataLoader(
        TensorDataset(X_train_t, y_train_t),
        batch_size=batch_size,
        shuffle=True
    )
    val_loader = DataLoader(
        TensorDataset(X_val_t, y_val_t),
        batch_size=batch_size,
        shuffle=False
    )

    if model_type == "linear":
        model = PrototypicalNetwork(input_size=X_train.shape[1], dropout = 0.1, output_size=num_superclasses)
    else:
        model = PrototypicalNetwork(input_size=X_train.shape[1], dropout = 0.1, output_size=num_superclasses)

    model = model.to(device)

    criterion = torch.nn.CrossEntropyLoss()
    optimizer = torch.optim.Adam(model.parameters(), lr=lr, weight_decay=1e-4)

    best_val_acc = -1.0
    best_state = None

    for epoch in range(epochs):
        model.train()
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)

            optimizer.zero_grad()
            logits = model(xb)
            loss = criterion(logits, yb)
            loss.backward()
            optimizer.step()

        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for xb, yb in val_loader:
                xb, yb = xb.to(device), yb.to(device)
                preds = model(xb).argmax(dim=1)
                correct += (preds == yb).sum().item()
                total += yb.size(0)

        val_acc = correct / max(total, 1)

        if val_acc > best_val_acc:
            best_val_acc = val_acc
            best_state = {k: v.cpu().clone() for k, v in model.state_dict().items()}

    model.load_state_dict(best_state)
    model = model.to(device)
    return model, best_val_acc

def evaluate_model(model, X_test, y_test_sc, device="cuda" if torch.cuda.is_available() else "cpu"):
    model.eval()
    X_test_t = torch.tensor(X_test, dtype=torch.float32).to(device)
    y_test_t = torch.tensor(y_test_sc, dtype=torch.long).to(device)

    with torch.no_grad():
        logits = model(X_test_t)
        preds = logits.argmax(dim=1)

    acc = (preds == y_test_t).float().mean().item()
    return acc, preds.cpu().numpy()
```

```python
def run_superclass_mlp(embeddings, labels, holdout_class, superclass_map=superclass_map, model_type="mlp"):
    X_train, y_train_fg, X_test, y_test_fg = make_split(embeddings, labels, holdout_class)

    y_train_sc, sc_to_id = make_superclass_ids(y_train_fg, superclass_map)
    y_test_sc = np.array([sc_to_id[superclass_map[int(y)]] for y in y_test_fg], dtype=np.int64)

    X_tr, X_val, y_tr, y_val = train_test_split(
        X_train, y_train_sc,
        test_size=0.2,
        random_state=42,
        stratify=y_train_sc)

    model, best_val_acc = train_superclass_classifier(
        X_tr, y_tr,
        X_val, y_val,
        num_superclasses=len(sc_to_id),
        model_type=model_type)

    test_acc, preds = evaluate_model(model, X_test, y_test_sc)
    return test_acc, best_val_acc, preds, sc_to_id
```

```python
x = l2_normalize(t1_euclidean_embeddings_known)
y = t1_euclidean_labels_known

for idx in range(0, 20):
    train_labels = y[y != idx]
    train_superclasses = {superclass_map[int(c)] for c in np.unique(train_labels)}
    test_superclass = superclass_map[idx]

    if test_superclass not in train_superclasses:
        print(f"skipping class {idx}: superclass '{test_superclass}' absent from training")
        continue

    test_acc, best_val_acc, preds, sc_to_id = run_superclass_mlp(x, y, idx)
    print(f"held-out class {idx}, test accuracy: {test_acc}, val accuracy: {best_val_acc}")
```
