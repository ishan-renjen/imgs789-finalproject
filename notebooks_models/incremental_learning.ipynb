```python
prototype_bank = dict()
replay_buffer = dict()

#set limits for replay buffer
num_samples = 1000
rep_frac = 0.7
hard_frac = 1 - rep_frac

#rng
seed=42

#shots
shot_size=20

#samples
rep_samples  = int(rep_frac * num_samples)
hard_samples = int(hard_frac * num_samples)

def l2_normalize(x):
    pca = PCA(n_components=16, random_state=42)
    class_embeddings_reduced = pca.fit_transform(x)
    return x / (np.linalg.norm(class_embeddings_reduced, axis=1, keepdims=True))

def cosine_loss(a, b, axis=-1):
    a = np.asarray(a, dtype=np.float64)
    b = np.asarray(b, dtype=np.float64)

    dot = np.sum(a * b, axis=axis)
    norm_a = np.linalg.norm(a, axis=axis)
    norm_b = np.linalg.norm(b, axis=axis)

    cos_sim = dot / (norm_a * norm_b)
    return 1.0 - cos_sim
```

```python
def rep_items(X, centers):
    chosen = []
    used = set()

    for center in centers:
        dists = np.linalg.norm(X-center, axis=1)
        order = np.argsort(dists)
        for idx in order:
            if idx not in used:
                used.add(idx)
                chosen.append(idx)
                break
    return np.array(chosen, dtype=int)

def select_representative_examples(class_emb):
    if len(class_emb) <= rep_samples:
        return class_emb.copy()

    k_eff = min(rep_samples, len(class_emb))
    km = KMeans(n_clusters=k_eff, random_state=seed, n_init=10)
    km.fit(class_emb)

    rep_idx = rep_items(class_emb, km.cluster_centers_)
    return class_emb[rep_idx]

def hard_scores(class_emb, class_num, prototypes):
    true_protos = prototypes[class_num]
    d_true = np.min(cosine_loss(class_emb[:, None, :], true_protos[None, :, :], axis=2),axis=1)

    d_wrong = np.full(len(class_emb), np.inf)

    for other_cls, other_protos in prototypes.items():
        if other_cls == class_num:
            continue
        d_cls = np.min(cosine_loss(class_emb[:, None, :], other_protos[None, :, :], axis=2),axis=1)
        d_wrong = np.minimum(d_wrong, d_cls)

    margins = d_wrong - d_true
    return margins

def select_hard_examples(class_emb, class_num, prototypes):
    if len(class_emb) <= hard_samples:
        return class_emb.copy()

    margins = hard_scores(class_emb, class_num, prototypes)
    hard_idx = np.argsort(margins)[:hard_samples]
    return class_emb[hard_idx]

def update_replay_buffer(support_emb, support_labels, prototype_bank, replay_buffer, class_list):
    new_buffer = dict(replay_buffer)

    for class_idx in class_list:
        class_emb = support_emb[support_labels == class_idx]
        if len(class_emb) == 0:
            continue

        rep = select_representative_examples(class_emb)

        if len(prototype_bank) > 1 and hard_samples > 0:
            hard = select_hard_examples(class_emb, class_idx, prototype_bank)
            combined = np.vstack([rep, hard]) if len(hard) > 0 else rep
        else:
            combined = rep

        combined = np.unique(combined, axis=0)

        if len(combined) > num_samples:
            combined = combined[:num_samples]

        new_buffer[class_idx] = combined

    return new_buffer
```

```python
def split_data(emb, labels):
    rng = np.random.default_rng(seed)

    support_x, support_y = [], []
    query_x, query_y = [], []

    for cls in np.unique(labels):
        idx = np.where(labels == cls)[0]
        rng.shuffle(idx)

        # keep at least 1 query example when possible
        n_support = min(shot_size, max(1, len(idx) - 1))
        support_idx = idx[:n_support]
        query_idx = idx[n_support:]

        support_x.append(emb[support_idx])
        support_y.append(labels[support_idx])

        if len(query_idx) > 0:
            query_x.append(emb[query_idx])
            query_y.append(labels[query_idx])

    support_x = np.concatenate(support_x, axis=0)
    support_y = np.concatenate(support_y, axis=0)
    query_x = np.concatenate(query_x, axis=0) 
    query_y = np.concatenate(query_y, axis=0)

    return support_x, support_y, query_x, query_y

def make_balanced_query_set(query_x, query_y, samples_per_class=None):
    rng = np.random.default_rng(seed)

    classes = np.unique(query_y)
    per_class_idx = {}

    for cls in classes:
        idx = np.where(query_y == cls)[0]
        print(len(idx))
        if len(idx) > 0:
            per_class_idx[cls] = idx

    if len(per_class_idx) == 0:
        return query_x, query_y

    min_count = min(len(idx) for idx in per_class_idx.values())

    if samples_per_class is None:
        n_take = min_count
    else:
        n_take = min(samples_per_class, min_count)

    chosen = []
    for cls, idx in per_class_idx.items():
        picked = rng.choice(idx, size=n_take, replace=False)
        chosen.append(picked)

    chosen = np.concatenate(chosen)
    rng.shuffle(chosen)

    return query_x[chosen], query_y[chosen]

def create_class_prototypes(support_emb, support_labels, replay_buffer, class_list, k):
    global prototype_bank

    for class_idx in class_list:
        current_mem = []

        if class_idx in replay_buffer and len(replay_buffer[class_idx]) > 0:
            current_mem.append(replay_buffer[class_idx])

        current_support_items = support_emb[support_labels == class_idx]
        if len(current_support_items) > 0:
            current_mem.append(current_support_items)

        support_2d = np.vstack(current_mem)

        k_means = KMedoids(n_clusters=k, metric='cosine', random_state=seed)
        k_means.fit(support_2d)
        prototype_bank[class_idx] = k_means.cluster_centers_

def predict_with_prototypes(query, prototype_bank):
    pred = None
    min_dist = float("inf")

    for class_idx, prototype_list in prototype_bank.items():
        dist = [cosine_loss(query, proto) for proto in prototype_list]
        min_class_dist = np.min(dist)

        if min_class_dist < min_dist:
            min_dist = min_class_dist
            pred = class_idx

    return pred, min_dist

def batch_predict(query_emb, prototype_bank):
    preds, dists = [], []
    for q in query_emb:
        p, d = predict_with_prototypes(q, prototype_bank)
        preds.append(p)
        dists.append(d)
    return np.array(preds), np.array(dists)
```

```python
def class_accuracy_table(y_true, y_pred):
    rows = []
    for cls in np.unique(y_true):
        mask = (y_true == cls)
        rows.append({"class_id": int(cls), "n_query": int(mask.sum()), "accuracy": float((y_pred[mask] == y_true[mask]).mean())})
    return pd.DataFrame(rows).sort_values(["accuracy", "n_query"], ascending=[True, False]).reset_index(drop=True)

def class_accuracy_table_with_superclass(y_true, y_pred, superclass_map):
    rows = []
    for cls in np.unique(y_true):
        mask = (y_true == cls)
        rows.append({"class_id": int(cls), "superclass": superclass_map[int(cls)], "n_query": int(mask.sum()), "accuracy": float((y_pred[mask] == y_true[mask]).mean())})
    return pd.DataFrame(rows)

def superclass_internal_variation_table(y_true, y_pred, superclass_map):
    class_df = class_accuracy_table_with_superclass(y_true, y_pred, superclass_map)
    out = (class_df.groupby("superclass", as_index=False).agg(num_classes=("class_id", "count"),
                                                              mean_class_acc=("accuracy", "mean"), 
                                                              std_class_acc=("accuracy", "std"), min_class_acc=("accuracy", "min"), 
                                                              max_class_acc=("accuracy", "max")).sort_values("std_class_acc", ascending=False).reset_index(drop=True))

    return out

def superclass_accuracy_table(y_true, y_pred, superclass_map):
    true_sc = np.array([superclass_map[int(x)] for x in y_true], dtype=object)
    pred_sc = np.array([superclass_map[int(x)] for x in y_pred], dtype=object)

    rows = []
    for sc in sorted(np.unique(true_sc)):
        mask = (true_sc == sc)
        rows.append({ "superclass": sc, "n_query": int(mask.sum()), "accuracy": float((pred_sc[mask] == true_sc[mask]).mean())})

    return pd.DataFrame(rows).sort_values(["accuracy", "n_query"], ascending=[True, False]).reset_index(drop=True)
```

```python
def run_incremental_fewshot(tasks, k=3):
    global prototype_bank
    prototype_bank = {}
    replay_buffer = {}
    
    seen_classes = []
    results = []
    per_task_details = []

    for task_id, (support_emb, support_labels, query_emb, query_labels, all_labels) in enumerate(tasks, start=1):
        task_classes = sorted(np.unique(all_labels).tolist())
        seen_classes = sorted(set(seen_classes).union(task_classes))

        create_class_prototypes(support_emb, support_labels, replay_buffer, seen_classes, k=k)

        preds, dists = batch_predict(query_emb, prototype_bank)
        acc = np.mean(preds == query_labels) if len(query_labels) > 0 else np.nan

        class_table = class_accuracy_table(query_labels, preds)
        superclass_table = superclass_accuracy_table(query_labels, preds, superclass_map)
        superclass_variation_table = superclass_internal_variation_table(query_labels, preds, superclass_map)

        results.append({"task": task_id, "accuracy": acc, "num_seen_classes": len(seen_classes),
                        "macro_class_acc": class_table["accuracy"].mean(), "class_acc_std": class_table["accuracy"].std(ddof=0),
                        "macro_superclass_acc": superclass_table["accuracy"].mean(), "superclass_acc_std": superclass_table["accuracy"].std(ddof=0)})

        per_task_details.append({"task": task_id, "y_true": query_labels.copy(), "y_pred": preds.copy(), "dists": dists.copy(),
                                 "class_table": class_table, "superclass_table": superclass_table, "superclass_variation_table": superclass_variation_table})

        replay_buffer = update_replay_buffer(support_emb=support_emb, support_labels=support_labels, prototype_bank=prototype_bank, replay_buffer=replay_buffer, class_list=task_classes)

    return results, replay_buffer, per_task_details

def test_incremental_fewshot(tasks):
    results = []
    per_task_details = []
    seen_classes = []

    for task_id, (emb, labels) in enumerate(tasks, start=1):
        preds, dists = batch_predict(emb, prototype_bank)

        acc = np.mean(preds == labels) if len(labels) > 0 else np.nan

        class_table = class_accuracy_table(labels, preds)
        superclass_table = superclass_accuracy_table(labels, preds, superclass_map)
        superclass_variation_table = superclass_internal_variation_table(labels, preds, superclass_map)

        results.append({"task": task_id, "accuracy": acc, "num_seen_classes": len(seen_classes),
                        "macro_class_acc": class_table["accuracy"].mean(), "class_acc_std": class_table["accuracy"].std(ddof=0),
                        "macro_superclass_acc": superclass_table["accuracy"].mean(), "superclass_acc_std": superclass_table["accuracy"].std(ddof=0)})

        per_task_details.append({"task": task_id, "y_true": labels.copy(), "y_pred": preds.copy(), "dists": dists.copy(),
                                 "class_table": class_table, "superclass_table": superclass_table, "superclass_variation_table": superclass_variation_table})

    return results, per_task_details
```

```python
def superclass_labels(y, superclass_map):
    return np.array([superclass_map[int(x)] for x in y], dtype=object)

def normalized_confusion(y_true, y_pred, labels):
    cm = confusion_matrix(y_true, y_pred, labels=labels)
    cm = cm.astype(float)
    row_sums = cm.sum(axis=1, keepdims=True)
    cm_norm = np.divide(cm, row_sums, out=np.zeros_like(cm), where=row_sums != 0)
    return pd.DataFrame(cm_norm, index=labels, columns=labels)

def superclass_confusion_table(y_true, y_pred, superclass_map):
    true_sc = superclass_labels(y_true, superclass_map)
    pred_sc = superclass_labels(y_pred, superclass_map)
    labels = sorted(np.unique(true_sc))
    return normalized_confusion(true_sc, pred_sc, labels)

def superclass_failure_breakdown(y_true, y_pred, superclass_map):
    true_sc = superclass_labels(y_true, superclass_map)
    pred_sc = superclass_labels(y_pred, superclass_map)

    rows = []
    for sc in sorted(np.unique(true_sc)):
        mask = (true_sc == sc)
        yt = y_true[mask]
        yp = y_pred[mask]
        psc = pred_sc[mask]

        total = len(yt)
        correct = (yp == yt)
        same_sc_wrong = (~correct) & (psc == sc)
        cross_sc_wrong = (psc != sc)

        wrong_dest = pd.Series(psc[cross_sc_wrong]).value_counts()
        top_wrong_sc = wrong_dest.index[0] if len(wrong_dest) > 0 else None
        top_wrong_sc_frac = float(wrong_dest.iloc[0] / total) if len(wrong_dest) > 0 else 0.0

        rows.append({"superclass": sc, "n_query": total, "acc": float(correct.mean()), "same_superclass_error_rate": float(same_sc_wrong.mean()),
                     "cross_superclass_error_rate": float(cross_sc_wrong.mean()), "top_wrong_superclass": top_wrong_sc, "top_wrong_superclass_frac": top_wrong_sc_frac})

    return pd.DataFrame(rows).sort_values("acc").reset_index(drop=True)

def within_superclass_confusion(y_true, y_pred, superclass_map, target_superclass):
    class_ids = sorted([cls for cls, sc in superclass_map.items() if sc == target_superclass])

    mask = np.array([superclass_map[int(y)] == target_superclass for y in y_true])
    yt = y_true[mask]
    yp = y_pred[mask]

    # keep only predictions that also land inside this superclass
    keep_cols = class_ids
    cm = confusion_matrix(yt, yp, labels=keep_cols).astype(float)

    row_sums = cm.sum(axis=1, keepdims=True)
    cm_norm = np.divide(cm, row_sums, out=np.zeros_like(cm), where=row_sums != 0)

    return pd.DataFrame(cm_norm, index=class_ids, columns=class_ids)

def top_confusion_pairs_within_superclass(y_true, y_pred, superclass_map, target_superclass, top_k=10):
    mask = np.array([superclass_map[int(y)] == target_superclass for y in y_true])
    yt = y_true[mask]
    yp = y_pred[mask]

    rows = []
    for t, p in zip(yt, yp):
        if t != p and superclass_map[int(t)] == target_superclass and superclass_map.get(int(p), None) == target_superclass:
            rows.append((int(t), int(p)))

    df = pd.DataFrame(rows, columns=["true_class", "pred_class"])
    return (df.value_counts().reset_index(name="count").sort_values("count", ascending=False).head(top_k).reset_index(drop=True))
```
