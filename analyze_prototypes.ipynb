```python
superclass_map = {
    0:'vehicle', 
    1:'vehicle',
    2:'outdoor', 
    3:'outdoor',
    4:'animal', 
    5:'animal', 
    6:'animal',
    7:'accessory',
    8:'sports', 
    9:'sports',
    10:'kitchen', 
    11:'kitchen',
    12:'food', 
    13:'food',
    14:'furniture', 
    15:'furniture',
    16:'electronic',
    17:'appliance', 
    18:'appliance',
    19:'indoor',
    #end task 1
    20:'person', 
    21:'vehicle',
    22:'vehicle', 
    23:'outdoor',
    24:'animal', 
    25:'animal', 
    26:'accessory',
    27:'sports',
    28:'sports', 
    29:'kitchen',
    30:'food', 
    31:'food',
    32:'food', 
    33:'furniture',
    34:'furniture', 
    35:'electronic',
    36:'electronic',
    37:'appliance', 
    38:'indoor',
    39:'indoor',
    #end task 2
    40: 'vehicle',
    41: 'vehicle',
    42: 'outdoor',
    43: 'animal',
    44: 'animal',
    45: 'animal',
    46: 'accessory',
    47: 'sports',
    48: 'sports',
    49: 'sports',
    50: 'kitchen',
    51: 'kitchen',
    52: 'food',
    53: 'food',
    54: 'kitchen',
    55: 'electronic',
    56: 'electronic',
    57: 'appliance',
    58: 'indoor',
    59: 'indoor',
    #end task 3
    60: 'vehicle',
    61: 'vehicle',
    62: 'outdoor',
    63: 'animal',
    64: 'animal',
    65: 'accessory',
    66: 'accessory',
    67: 'sports',
    68: 'sports',
    69: 'sports',
    70: 'kitchen',
    71: 'kitchen',
    72: 'food',
    73: 'food',
    74: 'food',
    75: 'kitchen',
    76: 'electronic',
    77: 'appliance',
    78: 'indoor',
    79: 'indoor',
}

def build_class_prototypes(embeddings, labels, k=3):
    prototypes = dict()

    for class_label in np.unique(labels):
        class_embeddings = embeddings[labels == class_label]

        if len(class_embeddings) <= k:
            prototypes[int(class_label)] = [class_embeddings.mean(axis=0)]
            continue

        kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
        kmeans.fit(class_embeddings)
        prototypes[int(class_label)] = [center for center in kmeans.cluster_centers_]

    return prototypes

def build_superclass_prototypes(class_prototypes: dict, superclass_map=superclass_map):
    superclass_prototypes = dict()

    for class_id, prototypes in class_prototypes.items():
        superclass = superclass_map[class_id]
        superclass_prototypes.setdefault(superclass, []).extend(prototypes)

    return superclass_prototypes

def predict_superclass(embedding, superclass_prototypes):
    pred_superclass = None
    lowest_dist = float('inf')

    for superclass, prototypes in superclass_prototypes.items():
        euclidean_dist = min(np.linalg.norm(embedding - prototype) for prototype in prototypes)
    
        if euclidean_dist < lowest_dist:
            lowest_dist = euclidean_dist
            pred_superclass = superclass
    
    return pred_superclass

def get_matched_prototypes(class_prototype_set1, class_prototype_set2):
    set1_array = [item for sublist in class_prototype_set1.values() for item in sublist]
    set2_array = [item for sublist in class_prototype_set2.values() for item in sublist]

    cost_matrix = cdist(set1_array, set2_array, metric='euclidean')
    row_ind, col_ind = scipy.linear_sum_assignment(cost_matrix)

    matched_pairs = list(zip(row_ind, col_ind))
    drift = cost_matrix[row_ind, col_ind].mean()

    return matched_pairs, drift
```
